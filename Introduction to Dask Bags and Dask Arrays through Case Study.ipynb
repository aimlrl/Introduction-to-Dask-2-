```python
import dask.dataframe as dd
from dask.diagnostics import ProgressBar
```

```python
df = dd.read_csv("./archive/Parking_Violations_Issued_-_Fiscal_Year_2017.csv")
```

```python
df
```

```python
missing_values = df.isnull().sum()
```

```python
missing_values
```

```python
missing_count = ((missing_values/df.index.size)*100)
```

```python
missing_count
```

```python
with ProgressBar():
    missing_count_val = missing_count.compute()
```

```python
missing_count_val
```

```python
columns_to_drop = missing_count_val[missing_count_val > 60].index
```

```python
columns_to_drop
```

```python
with ProgressBar():
        df_dropped = df.drop(columns_to_drop,axis=1).persist()
```

```python
df_dropped
```
